```python
# Changing the 'Latest_Launch' column to datetime

df["Latest_Launch"] = pd.to_datetime(df["Latest_Launch"])
print(df.dtypes)
```

```python
# Question 7
missing_values = df.isnull().sum
print(missing_values)
```

```python
# Question 8
df['__year_resale_value'].fillna(df['__year_resale_value'].mean())

# Drop rows with missing values in any other column
df.dropna(inplace=True)

# Check the result to ensure no missing values remain
print(df.isnull().sum())
```

```python
# Question 9
# Rename the column by removing the leading underscores
df.rename(columns={'__year_resale_value': 'year_resale_value'}, inplace=True)

print(df.columns)
```

```python
# Question 10
numerical_cols = df.select_dtypes(include=['float64', 'int64'])

# Calculate descriptive statistics
mean_values = numerical_cols.mean()
median_values = numerical_cols.median()
std_values = numerical_cols.std()

print("Mean values:\n", mean_values)
print("\nMedian values:\n", median_values)
print("\nStandard deviation values:\n", std_values)
```

Sales_in_thousands: The mean (59.11) and median (32.29) differ significantly, indicating a right-skewed distribution. There are likely a few cars with very high sales, pulling the mean higher than the median.

Horsepower: The mean181.2895) and standard deviation58.5970) show a wide variability in horsepower, indicating the cars in this dataset vary significantly in power, ranging from 55 to 450 HP.

Price_in_thousands: The me25.967.39) is higher than the medi21.662.80), suggesting that there are some high-priced cars inflating the mean.

Fuel_efficiency: The 24.11(23.84) and median(24.00) are very close, indicating a fairly symmetrical distribution for fuel efficiency. The values range between 15 and 45, showing some variation but not as much as in other columns.

```python
# Question 13
#calculate the mean of the 'Sales_in_thousands' column (replace 'Sales_in_thousands' if needed)
avg_sales_by_manufacturer = df.groupby('Manufacturer')['Sales_in_thousands'].mean()


# Get the manufacturer with the highest average sales
highest_avg_sales = avg_sales_by_manufacturer.idxmax()
highest_avg_sales_value = avg_sales_by_manufacturer.max()

# Get the manufacturer with the lowest average sales
lowest_avg_sales = avg_sales_by_manufacturer.idxmin()
lowest_avg_sales_value = avg_sales_by_manufacturer.min()

# Display the results
print(f"Manufacturer with the highest average sales volume: {highest_avg_sales} ({highest_avg_sales_value})")
print(f"Manufacturer with the lowest average sales volume: {lowest_avg_sales} ({lowest_avg_sales_value})")
```

```python
# Question 14
import matplotlib.pyplot as plt

car_prices = df['Price_in_thousands']

# 1. Descriptive statistics for car prices
print("Descriptive statistics for car prices:")
print(car_prices.describe())

# 2. Histogram to visualize the distribution
plt.figure(figsize=(10, 6))
plt.hist(car_prices, bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Car Prices')
plt.xlabel('Price (in thousands)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()


# 3. Box plot to identify the spread and any potential outliers
plt.figure(figsize=(8, 5))
plt.boxplot(car_prices, vert=False, patch_artist=True)
plt.title('Box Plot of Car Prices')
plt.xlabel('Price (in thousands)')
plt.show()
```

```python
# Question 20

# 1. Calculate the resale value ratio as a percentage of the initial price
df['resale_value_ratio'] = df['__year_resale_value'] / df['Price_in_thousands']

# 2. Find the car model with the highest resale value ratio
highest_resale_model = df.loc[df['resale_value_ratio'].idxmax()]

# 3. Display the result
print(f"The car model with the highest resale value ratio is:\n{highest_resale_model[['Model', 'Price_in_thousands', '__year_resale_value', 'resale_value_ratio']]}")
```

```python
# Question 21
# 1. Filter cars with engine size greater than 2.5 liters
filtered_cars = df[df['Engine_size'] > 2.5]

# 2. Sort the filtered cars by fuel efficiency in descending order
top_fuel_efficient_cars = filtered_cars.sort_values(by='Fuel_efficiency', ascending=False)

# 3. Select the top 3 most fuel-efficient cars
top_3_fuel_efficient = top_fuel_efficient_cars.head(3)

# Display the result
print("Top 3 most fuel-efficient cars with engine size above 2.5 liters:")
print(top_3_fuel_efficient[['Model', 'Engine_size', 'Fuel_efficiency']])
```

```python
# Question 22
# 1. Filter the dataset for Lexus models
lexus_cars = df[df['Manufacturer'] == 'Lexus']

# 2. Sort Lexus models by horsepower in descending order
lexus_sorted_by_horsepower = lexus_cars.sort_values(by='Horsepower', ascending=False)

# 3. Get the Lexus model with the highest horsepower
highest_horsepower_lexus = lexus_sorted_by_horsepower.head(1)

# Display the result
print("Lexus model with the highest horsepower:")
print(highest_horsepower_lexus[['Model', 'Horsepower']])
```
